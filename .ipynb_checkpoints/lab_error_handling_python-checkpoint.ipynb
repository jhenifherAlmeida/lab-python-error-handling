```python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        while True:
            try:
                quantity = int(input(f'Enter the quantity of {product}s available: '))
                if quantity < 0:
                    raise ValueError('Quantity cannot be negative.')
                inventory[product] = quantity
                break
            except ValueError as e:
                print(f'Error: {e}')   
    return inventory

def get_customer_orders(products, inventory):
    while True:
        try:
           num_orders = int(input('Enter the number of customer orders: '))
           if num_orders <= 0:
               raise ValueError('Number of orders must be a positive integer.')
           break 
        except ValueError as e:
            print(f'Error: {e}')
    print ('Enter the product names one by one:')           
    orders = set()
    i = 0
    while i < num_orders:
        product = input(f'Product {i+1}: ').strip().lower()
        if product not in products:
            print('Invalid product name. Please choose a product from the list.')
        elif inventory.get(product, 0) <= 0:
            print (f'Sorry, {product} is out of stock.')
        else:
            orders.add(product)
            i += 1
    return {'orders': orders}           

def calculate_total_price(customer_orders):
    prices = {} 
    for item in customer_orders['orders']:
        while True:
            try:
                price = float(input(f'enter the price of {item}: '))    
                if price < 0:
                    raise ValueError ('Price cannot be nagative.')   
                prices[item] = price
                break 
            except ValueError as e:
                print(f'Error: {e}')
    total_price = sum(prices.values())
    return total_price         

    


def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders['orders'])
    percentage_ordered = (total_products_ordered * 100) / len(products)
    return total_products_ordered, percentage_ordered

def print_order_statistics(order_statistics):
    print('\nOrder Statistics:')
    print(f'Total Products Ordered: {order_statistics[0]}')
    print(f'Percentage of unique products ordered: {order_statistics[1]:.1f}%')

def update_inventory(inventory, customer_orders):
    updated_inventory = {
        product: (inventory[product] - 1 if product in customer_orders['orders'] else inventory[product])
        for product in inventory
    }
    filtered_inventory = {product: qty for product, qty in updated_inventory.items() if qty > 0}
    return filtered_inventory

def print_updated_inventory(inventory):
    print('\nUpdated Inventory:')
    for product, qty in inventory.items():
        print(f'{product}: {qty}')



def main():
    products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']
    inventory = initialize_inventory(products)
    customer_orders = get_customer_orders(products, inventory)

    order_statistics = calculate_order_statistics(customer_orders, products)
    print_order_statistics(order_statistics)

    inventory = update_inventory(inventory, customer_orders)
    print_updated_inventory(inventory)

    total_price = calculate_total_price(customer_orders)
    print(f'\nTotal Price: {total_price:.1f}')



if __name__ == '__main__':
    main()
```
